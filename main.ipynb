```python
import networkx as nx
from random import *
import matplotlib as plt
```

```python
s = 5
G = nx.grid_graph(dim=[s,s])
nodes = list(G.nodes)
edges = list(G.edges)
p = []
for i in range(0, s):
    for j in range(0, s):
        p.append([i,j])

print(p)
```

```python
for i in range(0, len(nodes)):
    G.nodes[nodes[i]]['pos'] = p[i]
    G.nodes[nodes[i]]['val'] = randint(0,4)

print(G.nodes.data())
```

```python
nx.draw(G)
```

```python
list(nx.get_node_attributes(G,'val').values())
```

```python
nx.draw(G, node_color = list(nx.get_node_attributes(G,'val').values()), vmin=0, vmax=4, cmap = plt.cm.get_cmap('rainbow'))
```
