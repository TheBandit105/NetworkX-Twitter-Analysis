```python
import networkx as nx
from random import *
import matplotlib as plt
```

Create a 5x5 graph of nodes

```python
s = 5
G = nx.grid_graph(dim=[s,s])
nodes = list(G.nodes)
edges = list(G.edges)

nx.draw(G)
```

```python
for i in range(0, len(nodes)):
    G.nodes[nodes[i]]['val'] = randint(0,4)

print(G.nodes.data())
```

Let's get the 'val' property of each node

```python
list(nx.get_node_attributes(G,'val').values())
```

```python
nx.draw(G, node_color = list(nx.get_node_attributes(G,'val').values()), vmin=0, vmax=4, cmap = plt.cm.get_cmap('winter'))
```
